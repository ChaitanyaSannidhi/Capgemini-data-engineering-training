### From the column mixed_value, extract the numeric characters that appear at the very beginning of the value. The extraction must stop as soon as a non-numeric character appears.


```sql
SELECT 
    id,
    mixed_value,
    regexp_extract(mixed_value, '^[0-9]+', 0) AS starting_numbers
FROM regex_practice;
```

### From the column mixed_value, extract the numeric characters that appear at the very end of the value. No alphabetic or special characters should be included.


```sql
Select id , mixed_value , regexp_extract(mixed_value,'([0-9]+)$' , 0) as final_numbers from regex_practice
```

### From the column mixed_value, extract only the first single character of the value, regardless of whether it is a letter or a number.


```sql
Select id , mixed_value , regexp_extract(mixed_value,'^.',0) as first_character from regex_practice
```

### From the column mixed_value, extract only the last single character of the value.

```sql
Select id , mixed_value , regexp_extract(mixed_value,'.$',0) as last_letter from regex_practice
```

### From the column mixed_value, extract exactly two consecutive numeric characters that appearanywhere in the value.Do not extract more or fewer than two digits.


```sql
Select id , mixed_value , regexp_extract(mixed_value,'[0-9]{2}',0) as two_digits from regex_practice
```

### From the column mixed_value, extract exactly one numeric character that appears anywhere in the value.


```sql
Select id , mixed_value , regexp_extract(mixed_value,'[0-9]',0) as Numeric_digit from regex_practice
```

### From the column phone, extract the country code present at the beginning of the phone number. The extracted value must contain only the country code digits.


```sql
SELECT 
    id,
    phone,
    regexp_extract(phone, '^[+]?([0-9]+)-', 1) AS country_code
FROM regex_practice;
```

### From the column mixed_value, extract the numeric portion that is present between alphabetic characters.


```sql
SELECT 
    id,
    mixed_value,
    regexp_extract(mixed_value, '[a-zA-Z]+([0-9]+)[a-zA-Z]+', 1) AS numbers_between_letters
FROM regex_practice;
```

### From the column email, extract the text that appears before the at-symbol. The extracted value must not include the at-symbol itself.

```sql
select id , email , regexp_extract(email,'([a-zA-z0-9]+)@') as email_text from regex_practice
```

### From the column email, extract the text that appears after the at-symbol including the domain name.


```sql
select id , email , regexp_extract(email,'@(.+)',1) as email_text from regex_practice
```

### From the column email, extract only the domain name including the at-symbol.

```sql
select id , email , regexp_extract(email,'(@.+)',1) as email_domain from regex_practice
```

### From the column email, extract only the text that appears after the last dot in the email address.

```sql
SELECT 
    id,
    email,
    regexp_extract(email, '\\.([^.]+)$', 1) AS domain_extension
FROM regex_practice;
```
