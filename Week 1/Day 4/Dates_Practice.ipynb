### Calculating Weekday 

```sql
Select date_format(order_date,'EEEE') AS day_name , Weekday(order_date) as day from orders 
```

```sql
Select order_date ,date_format(order_date , 'dd-MM-yyyy') as new_date_format FROM orders 
```

```sql
SELECT 
customer_name,
order_amount,
NTILE(2) OVER(ORDER BY order_amount) AS spending_category
FROM orders;
```

```sql
Select customer_name , order_date , delivery_date , Date_diff(delivery_date , order_date) as Days_to_deliver from orders order by days_to_deliver
```

```sql
Select Year(order_date) as order_year,
Month(order_date) as order_month,
Day(order_date) as order_day,
date_format(order_date,'EEEE') AS day_name
from orders
```

```sql
Select date_format(order_date , 'EEEE') , dayofweek(order_date) as day_of_week from orders
```

```sql
Select date_format(order_date,'MMMM') as Month from orders
```

```sql
Select delivery_date , date_add(delivery_date , 10) as last_return_date from orders
```

```sql
Select delivery_date , add_months(delivery_date , 12) as warranty_period from orders
```

```sql
select date_format(order_timestamp,'HH:mm:ss') , date_format(order_timestamp + INterval 10 minutes , 'HH:mm:ss') as processing_time from orders
```

```sql
SELECT 
timestampdiff(HOUR, order_timestamp, current_timestamp()) AS hours,
timestampdiff(MINUTE, order_timestamp, current_timestamp()) % 60 AS minutes,
timestampdiff(SECOND, order_timestamp, current_timestamp()) % 60 AS seconds
FROM orders;
```

```sql
select last_day(order_date) as last_day_of_month from orders
```

```sql
SELECT 
order_date,
CASE
WHEN month(order_date) >= 4 
THEN concat(year(order_date),'-',year(order_date)+1)
ELSE concat(year(order_date)-1,'-',year(order_date))
END AS financial_year
FROM orders;
```
