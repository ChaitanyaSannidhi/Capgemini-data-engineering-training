### Count number of employees with NULL manager_id

```sql
Select Count(*) as no_manager from Employees1 where manager_id is null
```

### Replace NULL salary with 0

```sql
Select emp_id , name , ifnull(salary , 0) as salary from employees1
```

### Replace NULL bonus with 1000

```sql
Select emp_id , name , bonus , ifnull(bonus , 1000) as bonus_updated  from employees1
```

### Show order amount, if NULL replace with 500

```sql
Select customer_name , amount , ifnull(amount , 500) as amount_updated  from orders1
```

### Replace NULL stock with 0

```sql
SELECT product_id , stock , ifnull(stock , 0) as stock_updated FROM products
```

### Show employee earnings using: salary, if NULL use bonus


```sql
SELECT 
emp_id,
name,
salary ,
bonus ,
COALESCE(salary, bonus , 0) AS earnings
FROM employees1;
```

### Show product price: price → 1000 (default)


```sql
Select product_id , price , coalesce(price , 1000) as price_updated from products
```

### Get customer payment: amount → discount → 0


```sql
SELECT 
customer_name,
amount , 
discount ,
COALESCE(amount - discount ,amount, 0) AS customer_payment
FROM orders1;
```

### Calculate total earnings: salary + bonus (handle NULL properly)


```sql
Select salary , bonus , coalesce(salary + bonus , salary , bonus , 0) as Total_wage from employees1
```

## Replace empty discount (0) with NULL

```sql
SELECT 
discount,
NULLIF(discount, 0) AS discount_null
FROM orders1;
```

### Show final payable amount: amount - discount (handle NULL)


```sql
SELECT 
amount,
discount,
amount - COALESCE(discount,0) AS final_amount
FROM orders1;
```
