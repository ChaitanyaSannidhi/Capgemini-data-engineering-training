### Find employees who earn more than average salary

```sql
Select *  from employees21 where salary >(Select avg(salary) from employees21)
```

### Find employees who belong to the same department as 'Alice'

```sql
Select * from employees21 where department_id = (select department_id from employees21 where name = 'Alice')
```

## ### Get employees whose salary is equal to the minimum salary

```sql
Select * from employees21 where salary = (Select min(salary) from employees21)
```

### Find employees who have placed at least one order


```sql
SELECT *
FROM employees21 e
WHERE EXISTS (
    SELECT 1
    FROM orders21 o
    WHERE o.emp_id = e.emp_id
);
```

### Get employees whose salary is greater than ALL employees in department 102

```sql
SELECT * 
FROM employees21 
WHERE salary > (
    SELECT max(salary) 
    FROM employees21 
    WHERE department_id = 102
);
```

### Create a CTE to show employees with salary > 50,000 and fetch all records

```sql
WITH high_salary_employees AS (
    SELECT *
    FROM employees21
    WHERE salary > 50000
)

SELECT *
FROM high_salary_employees;
```

### Find average salary per department using CTE

```sql
with average_salary as (
  Select department_id ,avg(salary)
  from employees21
  group by department_id
)
select * from average_salary
```

### Use CTE to get employees and their department names

```sql
WITH emp_dept AS (
    SELECT 
        e.emp_id,
        e.name,
        e.salary,
        d.dept_name
    FROM employees21 e
    JOIN departments21 d
    ON e.department_id = d.dept_id
)

SELECT *
FROM emp_dept;
```

### Find total order amount per employee using CTE

```sql
With total_order_amount as(
  Select emp_id ,Sum(order_amount) as total_amount
  from orders21
  group by emp_id
)
select * from total_order_amount
```

```sql
WITH total_order_amount AS (
    SELECT 
        e.emp_id,
        e.name,
        SUM(o.order_amount) AS total_amount
    FROM employees21 e
    JOIN orders21 o
    ON e.emp_id = o.emp_id
    GROUP BY e.emp_id, e.name
)

SELECT *
FROM total_order_amount;
```

### Find employees whose salary is above department average using CTE

```sql
WITH dept_avg AS (
    SELECT department_id, AVG(salary) AS avg_salary
    FROM Employees21
    GROUP BY department_id
)
SELECT e.*
FROM Employees21 e
JOIN dept_avg d
ON e.department_id = d.department_id
WHERE e.salary > d.avg_salary;
```
